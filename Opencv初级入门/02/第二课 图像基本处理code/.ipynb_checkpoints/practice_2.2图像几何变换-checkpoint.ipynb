```python
import cv2
import numpy as np

img=cv2.imread('img2.png')
cv2.imshow('img',img)
cv2.waitKey(0)

#在X方向移动多少50距离，在Y方向移动多少25距离
H=np.float32([[1,0,50],[0,1,25]])
rows,cols=img.shape[:2]
print(img.shape)
print(rows,cols)

#先列后行
res=cv2.warpAffine(img,H,(cols,rows))
cv2.imshow('res',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img=cv2.imread('img2.png')
#方式一：通过设置缩放比例进行放大或缩小
#cv2.INTER_NEAREST（最近邻插值） cv2.INTER_AREA （区域插值） cv2.INTER_CUBIC（三次样条插值） cv2.INTER_LANCZOS4（Lanczos插值）
res1=cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)

cv2.imshow('origin_picture', img)
cv2.imshow('res1', res1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#方式2：通过直接设置图像大小
import cv2
import numpy as np

img=cv2.imread('img2.png')
height,width=img.shape[:2]
res2=cv2.resize(img,(int(0.2*width),int(0.2*height)),interpolation=cv2.INTER_LANCZOS4)

cv2.imshow('origin_picture', img)
cv2.imshow('res2', res2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img=cv2.imread('img2.png',1)
rows,cols=img.shape[:2]
#参数：旋转中心、旋转角度，旋转因子，1为逆时针
M=cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
print(M)
dst=cv2.warpAffine(img,M,(cols,rows))
while(True):
    cv2.imshow('img',img)
    cv2.imshow('dst',dst)
    if cv2.waitKey(1)&0xFF==27:
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#读取图片
img=cv2.imread('bird.png')
# 获取图像大小
rows,cols=img.shape[:2]
#图像放射变换矩阵
pos1 = np.float32([[50,50], [200,50], [50,200]])
pos2 = np.float32([[10,100], [200,50], [100,250]])
M=cv2.getAffineTransform(pos1,pos2)
print(M)
#图像放射变换
result=cv2.warpAffine(img,M,(2*cols,2*rows))
#显示图像
cv2.imshow('original',img)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('bird.png')
rows,cols=img.shape[:2]
#透视矩阵，变换前后的四个位置
pos1 = np.float32([[114, 82], [287, 156],
                   [8, 100], [143, 177]])
pos2 = np.float32([[0, 0], [188, 0], 
                   [0, 262], [188, 262]])
M=cv2.getPerspectiveTransform(pos1,pos2)
result=cv2.warpPerspective(img,M,(2*cols,2*rows))

cv2.imshow('original',img)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#encoding:utf-8
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('paper.png')
rows,cols=img.shape[:2]

#将图像高斯模糊化,去除噪声
img_gaus=cv2.GaussianBlur(img,(3,3),0)
#灰度化处理，方便得到图像的边缘
gray=cv2.cvtColor(img_gaus,cv2.COLOR_BGR2GRAY)
#边缘检测，检测图像边缘信息，边界阈值50，250
edges=cv2.Canny(gray,50,250,apertureSize = 3)
#通过霍夫变换，检测直线的函数得到A4纸边缘
lines = cv2.HoughLinesP(edges,1,np.pi/180,50,minLineLength=90,maxLineGap=10)
print(lines)
#输出4个顶点
for x1,y1,x2,y2 in lines[0]:
    print(x1,y1)
    print(x2,y2)
for x3,y3,x4,y4 in lines[1]:
    print(x3,y3)
    print(x4,y4)
# 绘制边缘
for x1,y1,x2,y2 in lines[0]:
    cv2.line(gray, (x1,y1), (x2,y2), (0,0,255), 1)
# 根据4个顶点设置透视变换矩阵
pos1 = np.float32([[114, 82], [287, 156], [8, 322], [216, 333]]) 
pos2 = np.float32([[0, 0], [188, 0], [0, 262], [188, 262]])
M = cv2.getPerspectiveTransform(pos1, pos2)
result = cv2.warpPerspective(img, M, (190, 272))

# 显示图像
cv2.imshow('img',img)
cv2.imshow('img_gaus',img_gaus)
cv2.imshow('gray',gray)
cv2.imshow('edges',edges)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#encoding:utf-8
import cv2  
import numpy as np
import matplotlib.pyplot as plt

#读取图片
img=cv2.imread('test2.png')
rows,cols,channel=img.shape
image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#图像平移
M = np.float32([[1, 0, 80], [0, 1, 30]])
rows, cols = image.shape[:2]
img1=cv2.warpAffine(image,M,(2*cols,2*rows))

#图像缩小
img2=cv2.resize(image,(200,100))

#图像放大
img3=cv2.resize(image,None,fx=1.5,fy=1.5)

#图像旋转
M=cv2.getRotationMatrix2D((rows/2,cols/2),30,1)
img4=cv2.warpAffine(image,M,(cols,rows))

#图像翻转
img5=cv2.flip(image,0)
img6=cv2.flip(image,1)

#图像的仿射
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
img7 = cv2.warpAffine(image, M, (rows,cols))

#图像的透射
pts1 = np.float32([[56,65],[238,52],[28,237],[239,240]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
M = cv2.getPerspectiveTransform(pts1,pts2)
img8 = cv2.warpPerspective(image,M,(200,200))

#循环显示图像
titles = [ 'source', 'shift', 'reduction', 'enlarge', 'rotation', 'flipX', 'flipY', 'affine', 'transmission']  
images = [image, img1, img2, img3, img4, img5, img6, img7, img8] 
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
plt.show()
```
