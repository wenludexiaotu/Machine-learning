```python
import cv2
import numpy as np

img=cv2.imread('girl2.png',cv2.IMREAD_UNCHANGED)
# 图像，目标图像通道数，卷积核尺寸，卷积核的数值
r=cv2.boxFilter(img,-1,(3,3),normalize=1)
d=cv2.boxFilter(img,-1,(3,3),normalize=0)
cv2.imshow('img',img)
cv2.imshow('r',r)
cv2.imshow('d',d)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('image/opencv.png')
#输入图像，卷积核大小，卷积核9个值最终除以9取平均
blur=cv2.blur(img,(3,3))
blur=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(121),plt.imshow(img),plt.title('original')
plt.xticks([]),plt.yticks([])#去掉坐标轴
plt.subplot(122),plt.imshow(blur),plt.title('blur')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('image/median.png')
#输入图像，卷积核大小，标准差越大越平滑
blur=cv2.GaussianBlur(img,(7,7),7)
blur=cv2.cvtColor(blur,cv2.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('image/median.png')
#输入图像，卷积核大小
blur=cv2.medianBlur(img,3)
blur=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('image/bilateral.png')
#输入图像;d=像损领域直径, 对于实时应用，建议取d=5;sigma1=灰度值相似性高斯函数标准差；sigma2=空间高斯函数标准差,一般大于10
blur=cv2.bilateralFilter(img,5,150,180)
blur=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np

def  custom_blur_demo(image):
    kernel=np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]], np.float32)#锐化
    dst=cv2.filter2D(image,-1,kernel=kernel)
    cv2.imshow('sharpen',dst)
img=cv2.imread('./image/sharpen.png')
cv2.imshow('img',img)
custom_blur_demo(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img = cv2.imread("./image/dark1.jpg")
B,G,R=cv2.split(img)
b=cv2.equalizeHist(B)
g=cv2.equalizeHist(G)
r=cv2.equalizeHist(R)
result=cv2.merge((b,g,r))
cv2.imshow('img',img)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img=cv2.imread('./image/dark1.jpg')
def adjust_gamma(image,gamma=1.5):
    g=1.0/gamma
    table=[]
    for i in range(256):
        table.append(((i/255.0)**g)*255)
    table=np.array(table).astype('uint8')
    return cv2.LUT(image,table)
img_gamma=adjust_gamma(img)
cv2.imshow("img",img)
cv2.imshow("img_gamma",img_gamma)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
