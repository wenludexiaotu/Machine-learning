```python
import cv2

img=cv2.imread('split.jpg')
cv2.imshow('img',img)
b,g,r=cv2.split(img)
cv2.imshow('b',b)
cv2.imshow('g',g)
cv2.imshow('r',r)

cv2.waitKey(0)
cv2.destroyAllWindows()
#可以看出b,g,r仍然是灰色，并不是红绿蓝
```

```python
import numpy as np
import cv2

image=cv2.imread('split.jpg')#读取图片
cv2.imshow('img',image)
cv2.waitKey(0)

B,G,R=cv2.split(image)
zeros=np.zeros(image.shape[:2],dtype='uint8')
cv2.imshow('BLUE',cv2.merge([B,zeros,zeros]))
cv2.imshow('GREEN',cv2.merge([zeros,G,zeros]))
cv2.imshow('RED',cv2.merge([zeros,zeros,R]))

cv2.waitKey(0)
cv2.destroyAllWindows()
#可以看出零填充后的三通道b,g,r仍然才是红绿蓝
# 单个通道显示的是该通道的颜色量，多个通道才会用滤色模式混合
print(image.shape,image.shape[:2])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img=cv2.imread('girl.jpg')
cv2.imshow('img',img)
cv2.waitKey(0)

img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap=plt.cm.gray)
hist=cv2.calcHist([img],[0],None,[256],[0,256])#图像，通道数、掩码、切分数量、像数值范围

plt.figure()
plt.title('hist')
plt.xlabel('Value')
plt.ylabel('Number')
plt.plot(hist)

plt.xlim([0,256])
plt.show()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

girl=cv2.imread('girl.jpg')
cv2.imshow('girl',girl)
cv2.waitKey(0)
color=['b','g','r']

for i,c in enumerate(color):
    hist=cv2.calcHist([girl],[i],None,[256],[0,256])
    plt.title('Hist')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(hist,color=c)
plt.show()
cv2.destroyAllWindows()
```
